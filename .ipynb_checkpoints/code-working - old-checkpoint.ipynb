```python
# Imports and dependencies
import pandas as pd
provisional_deaths = "resources/provisional_deaths.csv"
historical_deaths = "resources/causeofdeaththrough2020.csv"
```

```python
provisional_data = pd.read_csv(provisional_deaths)
prov_data_refined = provisional_data[["Year", "Month", "Intentional Self-Harm (Suicide)"]]
prov_data_years = prov_data_refined.groupby(["Year"])
prov_data_years_df = prov_data_years.sum()
prov_data_years_df = prov_data_years_df.rename(columns={"Intentional Self-Harm (Suicide)": "Deaths"})
prov_data_years_df = prov_data_years_df[["Deaths"]]
prov_data_years_df = prov_data_years_df.drop([2020, 2023])
prov_data_years_df["Population"] = ["332000000", "333000000"]
prov_data_years_df["Population"] = pd.to_numeric(prov_data_years_df["Population"])
prov_data_years_df["Crude Rate"] = ((prov_data_years_df["Deaths"] / prov_data_years_df["Population"]) * 100000)
prov_data_years_df["Crude Rate"] = pd.to_numeric(prov_data_years_df["Crude Rate"])


prov_data_years_df.index = prov_data_years_df.index.map(int)
prov_data_years_df
#prov_data_years_df = prov_data_years_df[["Intentional Self-Harm (Suicide)"]]
#prov_data_years_df.to_csv("output/provisional_deaths_by_year.csv", index=True)
```

```python
historical_data_df = pd.read_csv(historical_deaths)
historical_data_df = historical_data_df[["Year", "State", "Deaths", "Population", "Crude Rate"]]
historical_data_df = historical_data_df.dropna(how="any")
historical_data_df
```

```python
historical_year_state = historical_data_df.groupby(["Year"])
yearly_suicides_df = historical_year_state.sum()
yearly_suicides_df = yearly_suicides_df[["Deaths", "Population"]]
yearly_suicides_df.index = yearly_suicides_df.index.map(int)
yearly_suicides_df
```

```python
yearly_suicides_df["Crude Rate"] = ((yearly_suicides_df["Deaths"] / yearly_suicides_df["Population"]) * 100000)
yearly_suicides_df
#yearly_suicides_df.to_csv("output/yearly_suicide_rate_1999_2020.csv", index=True)
```
